### Plot losses of models

```python
tests_folder = "/Users/philvanlane/Documents/lc_ae/output/performance_tests/"

baseline_loss = np.load(tests_folder + "baseline/losses_per_epoch.npz")
unet_loss = np.load(tests_folder + "unet/losses_per_epoch.npz")
stellar_props_loss = np.load(tests_folder + "stellar_props/losses_per_epoch.npz")
lightweight_conv_loss = np.load(tests_folder + "lightweight_conv/losses_per_epoch.npz")
props_unet_loss = np.load(tests_folder + "props_unet/losses_per_epoch.npz")
long_curve_loss = np.load(tests_folder + "long_curve/losses_per_epoch.npz")
baseline_long_loss = np.load(tests_folder + "baseline_long/losses_per_epoch.npz")
cf_final_loss = np.load(tests_folder + "cf_final/losses_per_epoch.npz")
```

#### Plot

```python
plt.figure(figsize=(10,5))

plt.plot(baseline_loss['train_loss'],label = 'baseline (train)',color='black')
plt.plot(baseline_loss['val_loss'],color='black',linestyle='--',label='baseline(val)')

# plt.plot(unet_loss['train_loss'],label = 'extra channels (train)',color='firebrick')
# plt.plot(unet_loss['val_loss'],color='firebrick',linestyle='--',label='extra channels (val)')

# plt.plot(stellar_props_loss['train_loss'],label = 'stellar props',color='cyan')
# plt.plot(stellar_props_loss['val_loss'],color='cyan',linestyle='--')

# plt.plot(lightweight_conv_loss['train_loss'],label = 'lightweight conv',color='orange')
# plt.plot(lightweight_conv_loss['val_loss'],color='orange',linestyle='--')

plt.plot(props_unet_loss['train_loss'],label = 'props + unet',color='green')
plt.plot(props_unet_loss['val_loss'],color='green',linestyle='--')

plt.plot(long_curve_loss['train_loss'],label = 'props + unet, long curve',color='blue')
plt.plot(long_curve_loss['val_loss'],color='blue',linestyle='--')

plt.plot(baseline_long_loss['train_loss'],label = 'baseline, long curve',color='magenta')
plt.plot(baseline_long_loss['val_loss'],color='magenta',linestyle='--')

# plt.plot(cf_final_loss['train_loss'],label = 'cf_final',color='brown')
# plt.plot(cf_final_loss['val_loss'],color='brown',linestyle='--')


plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
