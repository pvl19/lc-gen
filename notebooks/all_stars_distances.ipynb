```python
from astroquery.vizier import Vizier
import pandas as pd
import numpy as np

Vizier.ROW_LIMIT = -1
```

```python
data_folder = '/Users/philvanlane/Documents/lc_ae/data/'
stars = pd.read_csv(data_folder + 'mocadb_add_TICS.csv')
gids = stars['GaiaDR3_ID'].tolist()
```

```python
all_stars['MG_quick'] = all_stars['G0'] + 5 * np.log10(all_stars['parallax']) - 10
```

```python
all_stars.to_csv(data_folder + 'phot_all.csv', index=False)
```

```python
result = Vizier.query_constraints(
        catalog="I/352/gedr3dis",
        Source=gids[0])
```

```python
df = pd.DataFrame(columns=result['I/352/gedr3dis'].colnames)
```

```python
df
```

```python

for i in range(len(gids)):
    if i % 500 == 0:
        print(f"Processing {i} of {len(gids)}")
    if len(df.loc[df['Source'] == gids[i]]) > 0:
        continue
    result = Vizier.query_constraints(
        catalog="I/352/gedr3dis",
        Source=gids[i]
    )
    if result.keys() == []:
        print(f"No result for {gids[i]}")
        continue
    df_record = result['I/352/gedr3dis'].to_pandas()
    df = pd.concat([df, df_record], ignore_index=True)
```

```python
df
```

```python
df.to_csv(data_folder + 'moca_add_bj_distances.csv', index=False)
```
