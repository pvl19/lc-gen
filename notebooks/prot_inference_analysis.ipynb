```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd

mpl.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Charter", "Bitstream Charter"],
    "font.style": "normal",
})
```

### Residuals Comparison

```python
prot_test_folder = '../output/prot_predictor/'

baseline = pd.read_csv(prot_test_folder + 'baseline_long-multiscale/kfold_predictions.csv')
with_conv = pd.read_csv(prot_test_folder + 'long_curve-multiscale/kfold_predictions.csv')

scatter_baseline = np.nanstd(baseline['log_residual'])
scatter_with_conv = np.nanstd(with_conv['log_residual'])
```

### Plot

```python
plt.figure(figsize=(8, 6))

# plt.hist(baseline['log_residual'], bins=50, alpha=0.5, label=r'baseline: $\sigma$ = {:.2f} dex'.format(scatter_baseline), color='black',density=True)
plt.hist(baseline['log_residual'], bins=50, label=r'without PS/ACF: $\sigma$ = {:.2f} dex'.format(scatter_baseline), color='magenta',density=True,
         histtype='step',edgecolor='teal',linewidth=3, alpha=0.8, linestyle='-')
plt.hist(with_conv['log_residual'], bins=50, label=r'with PS//ACF: $\sigma$ = {:.2f} dex'.format(scatter_with_conv), color='blue',density=True,
         histtype='step',edgecolor='goldenrod',linewidth=3, alpha=0.8, linestyle='-')

plt.xlabel('Log Residual', fontsize=12)
# plt.ylabel('Density', fontsize=12)
# plt.title('Prot Prediction Residuals', fontsize=14)
plt.axvline(0, color='k', linestyle='--', linewidth=1, alpha=0.5)
plt.legend()
plt.show()
```
